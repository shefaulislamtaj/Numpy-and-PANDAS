```python
import matplotlib.pyplot as plt
```

```python
data  = {
     'Sem 1' : 9.2,
     'Sem 2' : 6.6,
     'Sem 3' : 7.1,
     'Sem 4' : 9.4,
     'Sem 5' : 6.1,
     'Sem 6' : 9.5,
}
```

```python
sem =list(data.keys())
sgpa = list(data.values())
```

```python
fig =plt.figure(figsize=(10,5))
plt.bar(sem,sgpa, color='blue',width=0.3,
       #shadow = True,
       )
plt.xlabel('sem')
plt.ylabel('SGPA')
plt.show()
```
