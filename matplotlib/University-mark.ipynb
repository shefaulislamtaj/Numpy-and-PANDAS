```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
categories = ['Freshmen','sophomores','Juniors','Seniors']
values = np.array( [300,250,275,225])
colors = ['red', 'yellow','blue','green']
```

```python
plt.pie(values, labels = categories,
        colors = colors,
       autopct = '%1.1f%%',
       explode = [0,0,0,0.1],
       shadow = True,
       startangle =0,
       )
plt.title('Univercity')
plt.show()
```
