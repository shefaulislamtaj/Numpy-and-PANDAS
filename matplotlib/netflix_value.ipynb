```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
netflix_value = [540, 555, 530, 570, 600, 640, 610, 650, 670, 690, 710, 730]
```

```python
netflix_data = pd.DataFrame({
    'Month': months,
    'Value': netflix_value
})

# ==== Plot the Line Chart ====
plt.figure(figsize=(10, 6))
plt.plot(netflix_data['Month'], netflix_data['Value'],
         color='red', linewidth=2.5, marker='o', markersize=8, markerfacecolor='darkred')

# Add data labels
for i, value in enumerate(netflix_data['Value']):
    plt.text(i, value + 5, str(value), ha='center', fontsize=10, color='black')

# ==== Chart Titles and Labels ====
plt.title(" Netflix 2025 Sales/Value Trend", fontsize=16, fontweight='bold', color='red')
plt.xlabel("Month (2025)", fontsize=12)
plt.ylabel("Value (in USD millions)", fontsize=12)

# ==== Styling ====
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()

# ==== Show the Plot ====
plt.show()
```
